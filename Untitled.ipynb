```python
#line
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

# scatter
np.random.seed(123)
n = 500
X = np.random.normal(0, 0.5, n)
Y = np.random.normal(0, 0.5, n)
S = 25

#bar 
n1 = 16
X1 = np.arange(n1)
Y1 = (1-0.25*X1/n1) * np.random.uniform(0.25, 0.75, n1)
Y2 = (1-0.25*X1/n1) * np.random.uniform(0.25, 0.75, n1)

#image plot

def f(x,y): return (1-x/2+x**5+y**3)*np.exp(-x**2-y**2)
n2 = 128
x1 = np.linspace(-3,3,n2)
y1 = np.linspace(-3,3,n2)
X2,Y_2 = np.meshgrid(x1,y1)
Z = f(X2,Y_2)

#3d plot
X3,Y3= np.meshgrid(np.linspace(-1, 1, 25),
                     np.linspace(-1, 1, 25))
Z3 = np.sin(5*np.sqrt(X3**2+Y3**2))

# with subplot2grid

fig = plt.figure(figsize=(14, 10), dpi = 100) # initialize figure

ax = [None for _ in range(4)] # list to save many ax for setting parameter in each

ax[0] = plt.subplot2grid((2,3), (0,0), colspan=1)
ax[1] = plt.subplot2grid((2,3), (0,1), colspan=1)
ax[2] = plt.subplot2grid((2,3), (1,0), colspan=1)
ax[3] = plt.subplot2grid((2,3), (1,1), colspan=1)

#ax[4] = plt.subplot2grid((3,4), (1,3), colspan=1) #rowspan=2
#ax[5] = plt.subplot2grid((3,4), (0,2), colspan=1)


ax[0].scatter(X, Y, S/2, linewidth=0.00, facecolor="deepskyblue", edgecolor="None", alpha=0.5)
ax[0].scatter(X/2, Y/2, S/2, linewidth=0.00, facecolor="gold", edgecolor="None", alpha=0.2)

ax[1].plot(x, y , color="black", linewidth=2)
ax[1].plot(x, -y, color="blue", linewidth=2, linestyle = '--')

ax[2].bar(X1, +Y1, 1, facecolor='darkorange', edgecolor='black')
ax[2].bar(X1, -Y2, 1, facecolor='slateblue', edgecolor='black')

ax[3].imshow(Z, extent=[-1,+1,-1,+1], origin="upper", zorder=50,
              cmap = cmocean.cm.matter, interpolation="nearest")


for ix in range(4): 
    #ax[ix].set_title('ax[{}]'.format(ix)) # make ax title for distinguish:)
    ax[ix].set_xticks([]) # to remove x ticks
    ax[ix].set_yticks([]) # to remove y ticks
   
fig.tight_layout()
#plt.show()

plt.savefig('4_plots.png', bbox_inches = 'tight', transparent = True )
```

```python
X4,Y4= np.meshgrid(np.linspace(-1, 1, 25), np.linspace(-1, 1, 25))
Z4 = np.sin(5*np.sqrt(X4**2+Y4**2))

fig1 = plt.figure(figsize=(12,8), dpi = 100)
axe = fig1.add_subplot(111,  projection='3d')
im = axe.plot_surface(X4, Y4, Z4, rstride=1, cstride=1, alpha = 0.95, cmap=cmocean.cm.thermal, linewidth=.25) #rstride=1, cstride=1, vmin=-1.5, vmax=1, edgecolor = 'black'

cset = axe.contour(X4, Y4, Z4, zdir='z', offset = -1.05, linewidths = 0.3, cmap = "hot")
cset = axe.contour(X4, Y4, Z4, zdir='y',offset = -1.05, linewidths = 0.3, cmap = "hot")
cset = axe.contour(X4, Y4, Z4, zdir='x',offset = -1.05, linewidths = 0.3, cmap = "hot")

axe.set_xlim3d(np.min(X4),np.max(X4));
axe.set_ylim3d(np.min(Y4),np.max(Y4));
axe.set_zlim3d(np.min(Z4),np.max(Z4));

axe.set_xticks([])
axe.set_yticks([])
axe.set_zticks([])

fig1.colorbar(im, shrink=0.5, aspect=5, pad=0.001)
#fig1.tight_layout()
#plt.show()
axe.view_init(49, 50)
plt.savefig('threed.png', bbox_inches = 'tight', transparent = True)
```

```python
plt.figure(figsize=(12, 6), dpi = 200)

xxx = np.linspace(0.0, 100, 50)
yy = xxx*2
yyy = xxx*3
yyyy = xxx*4
yyyyy = xxx*5
yyyyyy = xxx*6
yyyyyyy = xxx*7
yyyyyyyy = xxx*8
yyyyyyyyy = xxx*9
yyyyyyyyyy = xxx*10

plt.plot(xxx, yy, linewidth = 0.3, linestyle = '--', label=r'$y=2x$')
plt.plot(xxx, yyy, linewidth = 0.5, linestyle = 'dotted', label=r'$y=3x$')
plt.plot(xxx, yyyy, linewidth = 1, linestyle = (0, (1,10)) , label=r'$y=4x$')
plt.plot(xxx, yyyyy, linewidth = 1.3, linestyle = 'dashdot', label=r'$y=5x$')
plt.plot(xxx, yyyyyy, linewidth = 1.5, linestyle = (0, (3, 1, 1, 1)), label=r'$y=6x$')
plt.plot(xxx, yyyyyyy, linewidth = 2, linestyle = (0, (3, 1, 1, 1, 1, 1)), label=r'$y=7x$')
plt.plot(xxx, yyyyyyyy, linewidth = 2.3, linestyle = (0, (1,1)), label=r'$y=8x$')
plt.plot(xxx, yyyyyyyyy, linewidth = 2.5, linestyle = (0, (3, 10, 1, 10)), label=r'$y=9x$')
plt.plot(xxx, yyyyyyyyyy, label = '$y=10x$')

#colour_blind_cycle = ['#377eb8', '#ff7f00', '#4daf4a',
                  #'#f781bf', '#a65628', '#984ea3',
                  #'#999999', '#e41a1c', '#dede00']
 
plt.xlim([0, 100])
plt.ylim([0, 1000])

current_handles, current_labels = plt.gca().get_legend_handles_labels()
# sort or reorder the labels and handles

reversed_handles = list(reversed(current_handles))
reversed_labels = list(reversed(current_labels))

# call plt.legend() with the new values
plt.legend(reversed_handles,reversed_labels, ncol = 2, fontsize = 'x-large', shadow=True)
# axis's labels
plt.xlabel(r'$X$-axis ($\dfrac{4\pi}{c}$)', fontsize = 18)
plt.ylabel(r'$Y$-axis ($\lambda$)', fontsize = 18)
# ticks 
plt.gca().tick_params(direction = 'in',labelsize = 14, width = 2)
# Make a plot with major ticks that are multiples of 20 and minor ticks that
# are multiples of 5.  Label major ticks with '.0f' formatting but don't label
# minor ticks.  The string is used directly, the `StrMethodFormatter` is
# created automatically.
plt.gca().xaxis.set_major_locator(MultipleLocator(20))

# For the minor ticks, use no labels; default NullFormatter.
plt.gca().xaxis.set_minor_locator(MultipleLocator(5))
# direction of minor ticks
plt.gca().tick_params(which="minor", direction="in")

#text
plt.text(60,650, r'This text starts at point (60,650)', bbox = {'facecolor':'yellow'}, horizontalalignment='right')
#graph title
plt.title(r'Graph example', fontsize = 20)

plt.savefig('example.png', bbox_inches = 'tight', transparent = True)
```
